## Creating DataFrame:

DataFrame can be created using various methods:

- **From dictionaries:** Each key-value pair in the dictionary corresponds to a column in the DataFrame.
- **From lists:** Lists can represent either rows or columns of the DataFrame.
- **From external data sources:** Data can be read from CSV, Excel, SQL databases, or other file formats.

```python
#Example 1: Creating DataFrame from dictionaries
data_dict = {
    'Name': ['John', 'Alice', 'Bob', 'Emily'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'Paris', 'London', 'Berlin']
}
df_dict = pd.DataFrame(data_dict)
df_dict
```

# Joining DataFrames:

Joining is used to combine columns from two doubtlessly different-listed DataFrames right into a single result
DataFrame. Here are a few examples:

### Inner Join: Based on Index

```python
#Creating two DataFrames
left = pd.DataFrame({"A": [1,2], "B": [3,4]}, index=['a', 'b'])
right = pd.DataFrame({"C": [5,6], "D": [7,8]}, index=['a', 'C'])

#Inner join based on index
result = left.join(right, how='inner')
result
```

### Left Join: Based on Index

```python
#Creating two DataFrames
left = pd.DataFrame({"A": [1,2], "B": [3,4]}, index=['a', 'b'])
right = pd.DataFrame({"C": [5,6], "D": [7,8]}, index=['a', 'C'])

#left join based on index
result = left.join(right, how='left')
result
```

# Merging DataFrames:

Merging is just like becoming a member of; however, it is extra versatile and lets in becoming a member of on
columns in addition to indexes. Here are some examples:

### Inner Merge

```python
#Creating two DataFrames
left = pd.DataFrame({"key": ['a','b'], "value": [1,2]})
right = pd.DataFrame({"key": ['a','c'], "value": [3,4]})

#Inner merge based on the 'key' column
result = pd.merge(left, right, on='key', how='inner')
result
```

### Left Merge:

```python
#Creating two DataFrames
left = pd.DataFrame({"key": ['a','b'], "value": [1,2]})
right = pd.DataFrame({"key": ['a','c'], "value": [3,4]})

#Left merge based on the 'key' column
result = pd.merge(left, right, on='key', how='left')
result
```

# Handling Missing Data in Pandas DataFrames:

## **dropna():**
This approach drops rows or columns containing missing values (NaN) from the DataFrame.
- axis=0 drops’ rows with NaN values.
- axis=1 drops columns with NaN values.

## **fillna():**
fillna() replaces missing values with certain values, along with a consistent or the imply of the data.
- **data.fillna(value)** fills all lacking values in the DataFrame with the specified value.
- **data.fillna(data.mean())** fills lacking values with the imply of each column.
- **data.fillna(approach=’pad’)** fills missing values with the preceding non-null cost.
- **data.fillna(method=’bfill’)** fills missing values with the next non-null value.

## **interpolate():**
interpolate() is used to interpolate lacking values within the DataFrame.
- **data.interpolate()** plays linear interpolation by means of default, filling NaN values with values
linearly interpolated among non-NaN values.
- Additional interpolation techniques may be specific the usage of the technique parameter, together
with ‘barycentric’, ‘pchip’, ‘akima’, ‘spline’, or ‘polynomial’.
- Interpolation can be in addition custom designed the use of non-obligatory arguments like **‘limit’, limit_direction, and limit_area** to control the interpolation behavior.

In simple terms, **interpolate()** in pandas is a way to fill in missing values (NaN) by estimating them based on the data you already have. It uses various methods to calculate the "best guess" for what the missing value should be.

**Linear Interpolation (Default)**

If you think of your data as a series of points on a graph, linear interpolation fills in the missing values by drawing a straight line between the known values and estimating where the missing values would fall along that line.

**Other Interpolation Methods**
The **method** parameter allows you to specify different ways of estimating missing values:

**method='polynomial' (Curved Lines)**
Fits a curved line through the data instead of a straight line.

**method='spline' (Smooth Curves)**
Fits a smooth curve through the data points.

**method='pad' (Forward Fill)**
Uses the previous value to fill the missing value (like fillna(method='pad')).

**method='barycentric' or method='pchip'**
These are more advanced techniques for estimating values using specific mathematical methods.

**Why Use Interpolation?**
- It's useful when you don't want to just fill missing values with constant numbers like 0 or the mean but want to maintain the trend or pattern in your data.
- It's especially common in time-series data, where missing values can be estimated based on time trends.

## **replace():**
replace() is used to replace values within the DataFrame.
- **data.replace(to_replace, value)** replaces values.
- It may be used to update precise NaN values or other values within the DataFrame.
- These techniques provide flexibility in coping with missing statistics in Pandas DataFrames, permitting
users to drop, fill, or interpolate lacking values based on their necessities.
- Additionally, **replace()** presents a way to update precise values, which include NaN, with desired values.


## Grouping:
Grouping entails splitting the information into organizations based on a few standards, which includes the
values of one or more columns. The **groupby()** feature in pandas is used to perform grouping. When you apply
**groupby()** to a DataFrame, it returns a GroupBy item, that is an intermediate step that allows you to carry out
diverse operations at the agencies.

Key points approximately grouping:
- Splitting: The information is break up into agencies primarily based on the specified criteria.
- Applying: A function is carried out to each group independently.
- Combining: The outcomes of the characteristic applied to every organization are mixed into a resulting facts shape.

```python
#Grouping by species and calculating the sum
#the result is a new table that shows the total (sum) of measurements (like sepal length, petal width, etc.) for each species of iris flower in your dataset.
iris.groupby('species').sum()
```

```python
#Grouping by species and calculating the mean
iris.groupby('species').mean()
```

```python
#iris data is divided into groups based on the species (e.g., Setosa, Versicolor, Virginica).
groupby = iris.groupby('species')
```

```python
# Looping over the group and printing the name and contents of each group
#In each iteration, name represents the species name (e.g., "Setosa"), and group represents the subset of the dataset that corresponds to that species.
#name: This represents the group label. When you group the data by species, each group gets a name corresponding to one of the species.
for name, group in groupby:
    print('\n' + name)
    print(group.head())
```

```python
# Applying aggregate function to the groupby object
#After the data is grouped, aggregate applies a specified function (in this case, "sum") to each group. 
#The "sum" function adds up the numeric values in each column for each group.
groupby.agg("sum")
```

```python
#Grouping by species with as_index set to False
sum_as_index_false = iris.groupby('species', as_index=False).sum()
sum_as_index_false
```

```python
#Applying size() method to the groupby object
#it counts how many records (rows) there are for each species in the dataset.
size_result = groupby.size()
size_result
```

```python
#Applying describe() method to the groupby object on a specific column
#summary statistics for the sepal_length column in each group. 
describe_result = groupby['sepal_length'].describe()
describe_result
```

```python
#Applying multiple methods to the groupby object
#The agg() function applies multiple aggregation functions to the sepal_length column for each species group.
multi_agg_result = groupby['sepal_length'].agg(["sum", "mean", "std"])
multi_agg_result
```

```python
#Applying a lambda function to the groupby object
#ddof=1 ensures that you're calculating the sample standard deviation, which is typically what is desired when working with a sample of data rather than a full population.
lambda_agg_result = groupby.agg(lambda x: x.std(ddof=1))
lambda_agg_result
```

```python
#Getting the largest and smallest values in each group
largest_result = groupby['sepal_length'].nlargest(3)
largest_result 
```

```python
smallest_result = groupby['petal_length'].nsmallest(3)
smallest_result 
```

```python
#Applying a custom function to the groupby object using apply()
#apply() is a flexible way to apply a function to each group in a groupby object. 
#custom_function takes a group (a series of petal_length values for one species) and returns a new DataFrame with two columns:
#original: The original petal_length values.
#demeaned: The demeaned values, which are calculated by subtracting the mean of the petal_length values for that species from each value in the group.
def custom_function(group):
    return pd.DataFrame({"original": group, "demeaned": group - group.mean()})

apply_result = groupby['petal_length'].apply(custom_function)
apply_result
```

```python
#Using qcut to create equal-sized buckets and then grouping by it
#qcut() is used to bin the sepal_length values into quantiles. The list [0, 0.25, 0.5, 0.75, 1] specifies the cut points for the quantiles:
#0 represents the lowest value (0th percentile).
#0.25 represents the 25th percentile (1st quartile).
#0.5 represents the 50th percentile (median, 2nd quartile).
#0.75 represents the 75th percentile (3rd quartile).
#1 represents the highest value (100th percentile).
#factor will be a categorical variable indicating the quantile bin each sepal_length value falls into.
#The groupby(factor) groups the data in the iris DataFrame by the factor created by qcut()
#Each group corresponds to one of the quantiles (from 0-25%, 25%-50%, 50%-75%, and 75%-100% of the sepal_length values).
#qcut_grouped.mean(): calculates the mean of all columns in the iris DataFrame (such as sepal_length, petal_length, sepal_width, petal_width) within each quantile group.
factor = pd.qcut(iris['sepal_length'], [0,0.25,0.5,0.75,1])
qcut_grouped = iris.groupby(factor, observed=False)
qcut_mean_result = qcut_grouped.mean(numeric_only=True)
qcut_mean_result
```
